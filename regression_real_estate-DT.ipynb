```python
import pandas
```

```python
data = pandas.read_csv("C:\\real_estate.csv",header=0)
```

```python
data
```

```python
from sklearn import tree
clf = tree.DecisionTreeRegressor()
```

```python
x=data.iloc[:,1:7] 
y=data.iloc[:,7]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
y_train, y_test, x_train, x_test = train_test_split(y,x,test_size=0.20)
arbre = clf.fit(x_train, y_train)
```

```python
y_pred = arbre.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
#Valeur de R²	Signification
#1.0	Modèle parfait : 100 % de la variance expliquée
#0.9 – 1.0	Excellente explication
#0.7 – 0.9	Bonne explication
#0.5 – 0.7	Moyenne, acceptable selon le domaine
#0 – 0.5	Faible explication (le modèle ne capture pas bien les tendances)
#< 0	Pire qu’un modèle qui prédit juste la moyenne
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import  numpy as np
def root_mean_squared_error(actual, predictions):
    return np.sqrt(mean_squared_error(actual, predictions))
```

```python
rmse=root_mean_squared_error(y_test, y_pred)
```

```python
rmse
```

```python
nrmse_range = rmse / (np.max(y_test) - np.min(y_test))
nrmse_mean  = rmse / np.mean(y_test)

print("RMSE :", rmse)
print("nRMSE (range) :", nrmse_range)
print("nRMSE (mean)  :", nrmse_mean)
```

```python
#nRMSE (normalisé par plage, moyenne)	Interprétation générale
#0.0 – 0.1	Excellent
#0.1 – 0.2	Très bon
#0.2 – 0.3	Acceptable / Bon
#0.3 – 0.5	Moyen (à améliorer)
#> 0.5	Faible / Mauvais
```
