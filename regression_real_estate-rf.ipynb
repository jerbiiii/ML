```python
import pandas
```

```python
data = pandas.read_csv("C:\\real_estate.csv",header=0)
```

```python
data
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=40)
```

```python
x=data.iloc[:,1:7] 
y=data.iloc[:,7]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
y_train, y_test, x_train, x_test = train_test_split(y,x,test_size=0.40)
```

```python
arbre = clf.fit(x_train, y_train)
```

```python
y_pred = arbre.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import  numpy as np
def root_mean_squared_error(actual, predictions):
    return np.sqrt(mean_squared_error(actual, predictions))
```

```python
root_mean_squared_error(y_test, y_pred)
```
